$\color{purple}{\text{Section 2: DataFrame Operations}}$

Description: This section illustrates various operations that can be performed on Pandas DataFrames, including arithmetic operations, aggregation functions, and applying functions.

DataFrame Operations:
- $\color{green}{\text{Arithmetic operations}}$: Addition, subtraction, multiplication, division.
- $\color{green}{\text{Aggregation functions}}$: sum(), mean(), std(), min(), max().
- $\color{green}{\text{Applying functions}}$: apply(), map(), applymap().

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Arithmetic operations: Addition, subtraction, multiplication, division.
df['C'] = df['A'] + df['B']
print("DataFrame after addition:\n", df)
```

```python
# Aggregation functions: sum(), mean(), std(), min(), max().
sum_value = df['A'].sum()
mean_value = df['A'].mean()
max_value = df['A'].max()
print("Sum value:", sum_value)
print("Mean value:", mean_value)
print("Max value:", max_value)
```

```python
# Applying functions: apply(), map(), applymap().
df['D'] = df['A'].apply(lambda x: x ** 2)
print("DataFrame after applying function:\n", df)
```

$\color{purple}{\text{Section 3: Indexing and Selecting Data}}$

Description: This section covers how to access and modify data within Pandas DataFrames using indexing and selecting techniques.

Indexing and Selecting Data:
- $\color{green}{\text{Accessing columns and rows.}}$
- $\color{green}{\text{Boolean indexing.}}$
- $\color{green}{\text{Using loc[] and iloc[].}}$

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Accessing columns and rows.
column_a = df['A']
row_0 = df.iloc[0] # 'i'loc: i stands for index
print("Column A:\n", column_a)
print("Row 0:\n", row_0)
```

```python
# Boolean indexing.
filtered_df = df[df['A'] > 1]
print("Filtered DataFrame:\n", filtered_df)
```

Differences between $\color{green}{\text{loc}}$ and $\color{green}{\text{iloc}}$
: **integer-location based indexing** is used to select rows and columns by their integer position (using index number).
- usage: works similarly to list slicing in Python
- syntax: `df.iloc[row_index,column_index]
- $\color{green}{\text{df.iloc[0]}}$: select first row
- $\color{green}{\text{df.iloc[0:3]}}$: select first 3 rows
- $\color{green}{\text{df.iloc[:,0]}}$: Select first column
- $\color{green}{\text{df.iloc[0,0]}}$: Select the element at the first row and first column

$\color{green}{\text{loc}}$: **label-based indexing** is used to select rows and columns by their labels (using row and column names)
- usage: it is more user friendly to select rows and columns by name
- syntax: 'df.loc[row_label,column_label]`
- $\color{green}{\text{df.loc['row1']}}$: Select the row with label 'row1'
- $\color{green}{\text{df.loc['row1':'row3']}}$: Select the rows from 'row1' to 'row3' (inclusive)
- $\color{green}{\text{df.loc[:,'column1']}}$: Select the column with label 'column1'
- $\color{green}{\text{df.loc['row1','column1']}}$: Select the element at the row 'row1' and column 'column1'

```python
# Using loc[] and iloc[].
loc_selection = df.loc[0:1, 'A']
iloc_selection = df.iloc[0:2, 0]
print("loc selection:\n", loc_selection)
print("iloc selection:\n", iloc_selection)
```

$\color{purple}{\text{Section 4: Data Manipulation}}$

Description: This section demonstrates how to manipulate data within Pandas DataFrames, including adding, deleting, and updating columns and rows.

Data Manipulation:
- $\color{green}{\text{Adding columns and rows.}}$
- $\color{green}{\text{Deleting columns and rows.}}$
- $\color{green}{\text{Updating values.}}$

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Adding columns and rows.
df['C'] = [7, 8, 9]
new_row = pd.DataFrame({'A': [4], 'B': [7], 'C': [10]})
df = pd.concat([df, new_row], ignore_index=True)
print("DataFrame after adding column and row:\n", df)
```

```python
# Deleting columns and rows.
df = df.drop('C', axis=1)
df = df.drop(0, axis=0)
print("DataFrame after deleting column and row:\n", df)
```

```python
# Updating values.
df.loc[1, 'A'] = 10
print("DataFrame after updating value:\n", df)
```

```python
# Create a df for 3 cars and 3 car prices
data = {
    'Car': ['Tesla Model 3','Tesla Model Y', 'BYD Seal'],
    'Price': [181000,191000,199800]
}
df = pd.DataFrame(data)
print(df)

# Add new column - country origin
df['origin'] = ['US','US','China']
print(df)


# Add new row: for a new car
new_car = pd.DataFrame({'Car':['Mazda CX3'],'Price':[115000],'origin':['Japan']})
df = pd.concat([df,new_car],ignore_index=True)
print(df)
```

```python
# Delete column 'origin'
df = df.drop('origin',axis=1)
print(df)
# Delete row: 'Tesla Model 3' from dataframe df
df = df.drop(0,axis=0)
print(df)
```

```python
# Update BYD Seal with a new car and new price
df.loc[2,'Car'] = 'BMW X5' # change BYD Seal to BMW X5
df.loc[2,'Price'] = 486800 # change 199800 to 486800
print(df)
```

$\color{purple}{\text{Section 5: Statistical Functions}}$

Description: This section introduces various statistical functions provided by Pandas, including descriptive statistics and correlation.

Statistical Functions:
- $\color{green}{\text{Descriptive statistics: describe(), mean(), median(), std().}}$
- $\color{green}{\text{Correlation: corr().}}$

```python
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [4, 5, 6, 7]})

# Descriptive statistics: describe(), mean(), median(), std().
desc_stats = df.describe()
mean_value = df['A'].mean()
median_value = df['A'].median()
std_value = df['A'].std()
print("Descriptive statistics:\n", desc_stats)
print("Mean value:", mean_value)
print("Median value:", median_value)
print("Standard deviation:", std_value)
```

```python
# Correlation: corr().
correlation = df.corr()
print("Correlation matrix:\n", correlation)
```

$\color{purple}{\text{Section 7: Handling Missing Values}}$

Description: This section demonstrates how to handle missing values in Pandas DataFrames.

Handling Missing Values:
- $\color{green}{\text{isna()}}$: Detect missing values.
- $\color{green}{\text{fillna()}}$: Fill missing values with a specified value.
- $\color{green}{\text{dropna()}}$: Remove rows or columns with missing values.

```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data_dict = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
             'Age': [25, np.nan, 35, 40, np.nan],
             'Salary': [50000, 60000, np.nan, 80000, 90000]}
df = pd.DataFrame(data_dict)

# Detecting missing values
print("Missing values:\n", df.isna())
```

```python
# Filling missing values
df_filled = df.fillna({'Age': df['Age'].mean(), 'Salary': 0})
print("DataFrame after filling missing values:\n", df_filled)
```

```python
# Dropping rows with missing values
df_dropped = df.dropna()
print("DataFrame after dropping rows with missing values:\n", df_dropped)
```

```python
# create a df with 5 students, 3 grades, 2 weights
import pandas as pd
import numpy as np
data = {
    'Student': ['George','Washington','William','Victor','Arthur'],
    'Grade': ['A',np.nan,np.nan,'A+','A+'],
    'Weight': [np.nan,60,70,np.nan,np.nan]
}
df = pd.DataFrame(data)
print(df)
```

```python
# detect missing values inside the df using .isna()
detect_missing = df.isna()
print("Missing values:\n", detect_missing)
```

```python
# Fill missing grades with 'U' and missing weight with 0
df_fill = df.fillna({'Grade':'U','Weight':0})
print("DataFrame treated for missing value:\n", df_fill)
```

```python
# Calculate the mean value of weights, calculate the mode of grade.
# Fill missing value of weights with mean weight, missing grade with mode grade
import numpy as np
data = {
    'Student': ['George','Washington','William','Victor','Arthur'],
    'Grade': ['A',np.nan,np.nan,'A+','A+'],
    'Weight': [np.nan,60,70,np.nan,np.nan]
}
df = pd.DataFrame(data)
print(df)
mean_weights = df['Weight'].mean()
# print(mean_weights)
mode_grade = df['Grade'].mode(dropna=True)
# print(mode_grade)

# use Pandas .values() to extract the string 'A+' from the dataframe'mode_grade'
grade = mode_grade.values[0]
print(grade)
df_fill = df.fillna({'Grade':grade,'Weight':mean_weights})
print(df_fill)
```
