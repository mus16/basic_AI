Customizing Plots:
- $\color {green}{\text {Changing colors.}}$

$\color{green}{\text { Section 3: Adding Labels and Titles}}$

Description: This section covers how to add labels, titles, and legends to Matplotlib plots.

Adding Labels and Titles:
- $\color{green}{\text {title()}}$: Add a title to the plot.
- $\color{green}{\text {xlabel(), ylabel()}}$: Add labels to the x and y axes.
- $\color{green}{\text {legend()}}$: Add a legend to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot with labels, title, and legend
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

$\color{green}{\text {  Section 4: Working with Multiple Plots}}$

Description: This section demonstrates how to create multiple plots within the same figure using subplots.

Working with Multiple Plots:
- $\color{green}{\text { subplot()}}$: Create multiple plots in the same figure.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()

# First subplot
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("Sin(x)")

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Cos(x)")

plt.tight_layout()
plt.show()
```

$\color{green}{\text{Section 3: Adding Labels and Titles}}$

Description: This section covers how to add labels, titles, and legends to Seaborn plots.

Adding Labels and Titles:
- set_title():Add a title to the plot.
- set_xlabel(), set_ylabel(): Add labels to the x and y axes.
- legend(): Add a legend to the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Line plot with labels, title, and legend
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
sns.lineplot(x=x, y=y1, label='sin(x)')
sns.lineplot(x=x, y=y2, label='cos(x)')
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

Exercise Adding Labels and Titles

- Create a numpy, x of number from 0 to 10 with 200 elements
- Create y1 which is sin(x)
- Create y2 which is sin(x+50)
- Plot both graphs with labels: sin(x) and sin(x+50)
- Title: Sine Trigonometric Function: Phase Shift
- xlabel: x, ylabel:value
- make sure legend is inside the graph

```python

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

x = np.linspace(0,10,200)
y1 = np.sin(x)
y2 = np.sin(x + 50)
plt.figure()
sns.lineplot(x=x,y=y1,label='sin(x)')
sns.lineplot(x=x,y=y2,label='sin(x+50)')
plt.title('Sine Trigonometric Function: Phase Shift')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.show()
```

$\color{green} {\text {Section 4: Working with Multiple Plots}}$

Description: This section demonstrates how to create multiple plots within the same figure using Seaborn's FacetGrid and Matplotlib's subplots.

Working with Multiple Plots:
- $\color{green} {\text {FacetGrid()}}$: Create multiple plots in the same figure.
- $\color{green} {\text {subplot()}}$: Create multiple plots in the same figure.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Multiple plots using FacetGrid
tips = sns.load_dataset("tips")
print(tips.head())
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

$\color {green}{\text{ Section 5: Creating Advanced Plots}}$

Description: This section covers advanced plots such as histograms, box plots, and heatmaps using Seaborn.

Creating Advanced Plots:
- $\color {green}{\text{ histplot()}}$: Create histograms.
- $\color {green}{\text{ boxplot()}}$: Create box plots.
- $\color {green}{\text{ heatmap()}}$: Create heatmaps.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Histogram
data = np.random.randn(1000)
plt.figure()
sns.histplot(data, bins=30, kde=True)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot
data = [np.random.randn(100) for _ in range(4)]
plt.figure()
sns.boxplot(data=data)
plt.title("Box Plot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# Heatmap
data = np.random.rand(10, 10)
plt.figure()
sns.heatmap(data, cmap='viridis')
plt.title("Heatmap")
plt.show()
```

$\color{green}{\text{Matplotlib VS Seaborn}}$

- $\color{green}{\text{Matplotlib}}$: Great for **detailed and customized plots**; requires more code and provides extensive control over plot appearance.
- $\color{green}{\text{Seaborn}}$: Simplifies the creation of complex statistical plots; offers attractive default aesthetics and integrates well with Pandas for **quick data visualization**.
